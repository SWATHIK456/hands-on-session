```python
sales_by_date=df.groupby("Order Date")['Sales'].sum()
sales_by_date.head()
```

```python
sales_by_date.index
```

```python
plt.figure(figsize=(20,5))
sales_by_date=df.groupby("Order Date")['Profit'].sum()
sales_by_date.head()
plt.plot(sales_by_date.index,sales_by_date.values)
plt.title("sale by index",fontsize=20)
plt.xlabel("order date",fontsize=14)
plt.ylabel("Profit",fontsize=14)
plt.title("daily profit trend")
plt.xticks(rotation=45)
plt.grid(True,linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(20,5))
sales_by_date=df.groupby("Region")['Profit'].sum()
plt.bar(sales_by_date.index,sales_by_date.values,color='g')
plt.title("Total slaes by region",fontsize=20)
plt.xlabel("Region",fontsize=14)
plt.ylabel("sales",fontsize=14)
#plt.title("daily profit trend")
#plt.xticks(rotation=45)
#plt.grid(True,linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sales_by_date=df.groupby("Region")['Profit'].sum().sort_values(ascending=False)
plt.bar(sales_by_date.index,sales_by_date.values,color='g')
plt.title("Total slaes by region",fontsize=20)
plt.xlabel("Region",fontsize=14)
plt.ylabel("sales",fontsize=14)
#plt.title("daily profit trend")
#plt.xticks(rotation=45)
#plt.grid(True,linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sales_by_date=df.groupby("Region")['Profit'].sum().sort_values(ascending=False)
plt.bar(sales_by_date.index,sales_by_date.values,color=['g','r','b','orange','violet','indigo','yellow','pink'])
plt.title("Total slaes by region",fontsize=20)
plt.xlabel("Region",fontsize=14)
plt.ylabel("sales",fontsize=14)
```

```python
plt.figure(figsize=(5,5))
sales_by_date=df.groupby("Region")['Profit'].sum().sort_values(ascending=False)
plt.barh(sales_by_date.index,sales_by_date.values,color=['g','r','b','orange','violet','indigo','yellow','pink'])
plt.title("Total slaes by region",fontsize=20)
plt.xlabel("Region",fontsize=14)
plt.ylabel("sales",fontsize=14)
plt.xticks(rotation=45)
```
